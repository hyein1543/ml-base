# 데이터 분리

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, shuffle=True, random_state=SEED)
```

```python
train.head()
```

```python
X_train = train.drop(["car_name","mpg"], axis=1) # 쓸모없는 컬럼과 타겟컬럼 빼기
X_train.head()
```

```python
y_train = train["mpg"]
y_train.head()
```

# 데이터 전처리

```python
# 결측치
X_train.isna().sum()
```

```python
X_train.dropna().isna().sum() # 결측치 그냥 제거하는 방법
```

```python
X_train["horsepower"].mean(), X_train["horsepower"].median() # 결측치를 평균값을 넣을지, 중앙값을 넣을지
```

```python
sns.histplot(X_train["horsepower"])
```

```python
horsepower_median = X_train["horsepower"].median()

X_train["horsepower"] = X_train["horsepower"].fillna(X_train["horsepower"].median())
```

```python
X_train.isna().sum(), X_train.shape
```

### 수동계산

```python
mean = np.mean(X_train,axis=0)
mean # 이 값은 그대로 저장했다가 test데이터에도 처리해야하고, 서비스를 제공할 때도 함께 제공해야함
```

```python
std = np.std(X_train,axis=0)
std # 이 값은 그대로 저장했다가 test데이터에도 처리해야하고, 서비스를 제공할 때도 함께 제공해야함
```

```python
(X_train - mean)/std
```

### 스케일러 사용

```python
from sklearn.preprocessing import StandardScaler

xss = StandardScaler()
#ss.fit(X_train)  # 평균,표준편차 계산
#ss.transform(X_train) # 적용

X_train_s = xss.fit_transform(X_train) # 스케일러 함수 쓰면 데이터 타입이 numpy로 바뀜
X_train_s
```

```python
xss.mean_, xss.var_
```

```python
y_train.to_numpy().reshape(318,1).shape
```

```python
# y도 스케일링 해줘야함. 연비가 10단위인데 RMSE 구할 때 값이 커질 수 있음
yss = StandardScaler()
y_train_s = yss.fit_transform(y_train.to_numpy().reshape(318,1)) # 함수에서 2차원 이상만 처리할 수 있도록 함. np로 2차원으로 바꿔주기
y_train_s = y_train_s.reshape(-1)
y_train_s[:10]
```

## 최종 학습데이터

# 모델 (베이스라인)

```python
from sklearn.linear_model import LinearRegression # 내부적으로 최소제곱법으로 돌아감

model = LinearRegression()
model.fit(X_train_s, y_train_s)
```

```python
model.coef_, model.intercept_ # 절편
```

```python
X_test = test.drop(["car_name","mpg"], axis=1) # 쓸모없는 컬럼과 타겟컬럼 빼기
y_test = test["mpg"]

X_test.shape, y_test.shape
```

## 스케일링



```python
X_test_s = xss.transform(X_test) # fit_transform 사용X-> 그럼 테스트 데이터에서 다시 mean, std 계산
```

```python
y_test_s = yss.transform(y_test.to_numpy().reshape(-1,1)).reshape(-1)
y_test_s[:10]
```

```python
y_test_s[:10]
```

```python
# 스케일링한거 원본으로 돌리기
y_pred = yss.inverse_transform(y_pred_s.reshape(-1,1)).reshape(-1)
y_pred[:10]
```

# 평가

```python
from sklearn.metrics import mean_squared_error, r2_score

def print_metrics(y_true, y_pred):
  mse = mean_squared_error(y_true, y_pred)
  r2  = r2_score(y_true, y_pred)
  print(f"MSE : {mse}, R2 : {r2}")

print_metrics(y_test, y_pred)
```

# Ridge

```python
y_pred_s = ridge.predict(X_test_s)
y_pred = yss.inverse_transform(y_pred_s.reshape(-1,1)).reshape(-1)
```

```python
print_metrics(y_test, y_pred)
```

# Lasso

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(X_train_s, y_train_s)
```

```python
y_pred_s = lasso.predict(X_test_s)
y_pred = yss.inverse_transform(y_pred_s.reshape(-1,1)).reshape(-1)
```

```python
print_metrics(y_test, y_pred)
```
